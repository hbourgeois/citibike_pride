```python
#import dependencies
import os
import csv
import pandas as pd
```

```python
#read in the csv file
pride_2023=pd.read_csv('pride_2023.csv')
pride_2023.head()
```

```python
#convert the csv file to a dataframe
pride_2023_df=pd.DataFrame(pride_2023)

#convert started_at and ended_at to datetime
pride_2023_df['started_at'] = pd.to_datetime(pride_2023_df['started_at'])
pride_2023_df['ended_at'] = pd.to_datetime(pride_2023_df['ended_at'])
```

```python
#split the started_at and ended_at columns into date and time
pride_2023_df['start_date'] = pride_2023_df['started_at'].dt.date
pride_2023_df['start_time'] = pride_2023_df['started_at'].dt.time
pride_2023_df['end_date'] = pride_2023_df['ended_at'].dt.date
pride_2023_df['end_time'] = pride_2023_df['ended_at'].dt.time

#drop the started_at and ended_at columns
pride_2023_df=pride_2023_df.drop(columns=['started_at','ended_at'])
```

```python
#create a new column for the duration of the event
pride_2023_df['duration'] = (pd.to_datetime(pride_2023_df['end_time'].astype(str)) - pd.to_datetime(pride_2023_df['start_time'].astype(str)))

#convert the duration to minutes
pride_2023_df['duration'] = pride_2023_df['duration'].dt.total_seconds()/60

#convert the duration to an integer
pride_2023_df['duration'] = pride_2023_df['duration'].astype(int)
```

```python
#display the dataframe
pride_2023_df.head()
```

```python
#change the name of the * values in the rideable_type column to "next_gen"
pride_2023_df['rideable_type'] = pride_2023_df['rideable_type'].replace(['*'],'next_gen')

#display the dataframe
pride_2023_df.head()
```

```python
#drop unnecessary columns
pride_2023_df=pride_2023_df.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'])
```

```python
#export the dataframe to a new csv file
pride_2023_df.to_csv('pride_2023_duration.csv', index=False)
```
